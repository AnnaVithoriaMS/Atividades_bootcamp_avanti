```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_pinguins = sns.load_dataset('penguins')
df_pinguins.head()
```

```python
df_pinguins.info()
```

```python
sns.set_style('whitegrid')
ax = sns.histplot(df_pinguins, x = 'body_mass_g')

plt.ylabel('Quantidades')
plt.xlabel('Massa dos pinguins')
ax.set(frame_on=False)
ax.axhline(0,color = 'k', clip_on=False)
plt.show()
```

```python
df_gorjeta = sns.load_dataset('tips')
df_gorjeta.head()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(6,3))
ax = sns.boxplot(data = df_gorjeta, x = 'day', y='total_bill', color='pink')
plt.ylabel('Valor em dolares')
plt.xlabel('Dia da semana')
plt.title('Pagamento de gorjetas por dias da semana')
plt.show()
```
